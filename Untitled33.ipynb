```python
from datasets import load_dataset

# Load your dataset
dataset = load_dataset("sherry41891/chitrali_dataset")
```

```python
dataset
```

```python
from datasets import DatasetDict
from PIL import Image
from io import BytesIO

train_dataset = dataset["train"]

# Define function to fix images
def load_real_image(example):
    if isinstance(example['image'], str):
        # Load from file path
        image = Image.open(example['image']).convert('RGB')
    elif isinstance(example['image'], list):
        # If it's a list of ints (bytes)
        image_bytes = bytes(example['image'])
        image = Image.open(BytesIO(image_bytes)).convert('RGB')
    else:
        raise ValueError(f"Unknown image type: {type(example['image'])}")

    return {
        "image": image,
        "caption": example["caption"]
    }

# Map it over train split
fixed_train = train_dataset.map(load_real_image)

# Now recreate the DatasetDict
fixed_dataset_dict = DatasetDict({
    "train": fixed_train
})

# Done ✅
print(fixed_dataset_dict)
print(fixed_dataset_dict['train'][0]['image'])  # should be PIL.Image
```

```python
def load_real_image(example):
    img = example['image']

    if isinstance(img, dict) and 'bytes' in img:
        # Huggingface format: {'bytes': b'...'}
        image = Image.open(BytesIO(img['bytes'])).convert('RGB')

    elif isinstance(img, str):
        # Image path
        image = Image.open(img).convert('RGB')

    elif isinstance(img, list):
        # list of ints
        image_bytes = bytes(img)
        image = Image.open(BytesIO(image_bytes)).convert('RGB')

    else:
        raise ValueError(f"Unknown image type: {type(img)}")

    return {
        "image": image,
        "caption": example["caption"]
    }


fixed_train = train_dataset.map(load_real_image)

fixed_dataset_dict = DatasetDict({
    "train": fixed_train
})

# Test
print(fixed_dataset_dict["train"][0]["image"])
print(fixed_dataset_dict["train"][0]["caption"])
```
